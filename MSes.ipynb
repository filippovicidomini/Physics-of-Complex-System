```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load data
kb = 8.617e-5  # eV/K
N = 1000

Np = np.linspace(0, N)
```

```python
def entropy(Np, N):
    return kb * (N * np.log(N) - Np * np.log(Np) - (N - Np) * np.log(N - Np))
```

```python
max = entropy(N / 2, N)
max_calc = kb * N * np.log(2)
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig, ax = plt.subplots()
ax.hlines(max, 0, N, label="max entropy")
ax.hlines(max_calc, 0, N, label="max entropy calc", colors="red")
plt.plot(Np, entropy(Np, N))
plt.show()
```
